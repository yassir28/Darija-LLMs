```python
from nltk.corpus import wordnet as wn
import pandas as pd
df = pd.read_csv('verbs/verbs.csv')

df = df.head(953)
eng = df['eng']
eng = [verb.replace(" ", "_") for verb in eng]


relevant = verbs['relevant']
eng_relevant = verbs['eng_relevant']
persona = verbs['persona']
eng_persona = verbs['eng_persona']
delimiters = verbs['delimiters']
eng_delimiters = verbs['eng_delimiters']
fewshot = verbs['few-shot']
eng_fewshot = verbs['eng_few-shot']



p1_transl = [verb.replace(" ", "_") for verb in relevant]
eng_p1_transl = [verb.replace(" ", "_") for verb in eng_relevant]
p2_transl = [verb.replace(" ", "_") for verb in persona]
eng_p2_transl = [verb.replace(" ", "_") for verb in eng_persona]
p3_transl = [verb.replace(" ", "_") for verb in delimiters]
eng_p3_transl = [verb.replace(" ", "_") for verb in eng_delimiters]
p5_transl = [verb.replace(" ", "_") for verb in fewshot]
eng_p5_transl = [verb.replace(" ", "_") for verb in eng_fewshot]


def max_path_similarity(word1, word2):
    synsets1 = wn.synsets(word1, pos=wn.VERB)
    synsets2 = wn.synsets(word2, pos=wn.VERB)
    max_score = 0
    for syn1 in synsets1:
        for syn2 in synsets2:
            score = syn1.path_similarity(syn2)
            if score and score > max_score:
                max_score = score
    return max_score

def evaluate_pairs(transl,threshold=0.5):
    results = []
    for i in range(len(transl)):
        word1, word2 = transl[i], eng[i]
        similarity = max_path_similarity(word1, word2)
        results.append((word1, word2, similarity, similarity >= threshold))
    return results



results_p1 = evaluate_pairs(p1_transl)
results_eng_p1 = evaluate_pairs(eng_p1_transl)
results_p2 = evaluate_pairs(p2_transl)
results_eng_p2 = evaluate_pairs(eng_p2_transl)
results_p3 = evaluate_pairs(p3_transl)
results_eng_p3 = evaluate_pairs(eng_p3_transl)
results_p5 = evaluate_pairs(p5_transl)
results_eng_p5 = evaluate_pairs(eng_p5_transl)


verbs['similarity_p1'] = [item[2] for item in results_p1]
verbs['similarity_eng_p1'] = [item[2] for item in results_eng_p1]
verbs['similarity_p2'] = [item[2] for item in results_p2]
verbs['similarity_eng_p2'] = [item[2] for item in results_eng_p2]
verbs['similarity_p3'] = [item[2] for item in results_p3]
verbs['similarity_eng_p3'] = [item[2] for item in results_eng_p3]
verbs['similarity_p5'] = [item[2] for item in results_p5]
verbs['similarity_eng_p5'] = [item[2] for item in results_eng_p5]


vp1_s = verbs['similarity_p1']
vep1_s = verbs['similarity_eng_p1']
vp1= verbs['similarity_p1'].mean()
vep1= verbs['similarity_eng_p1'].mean()

vp2_s = verbs['similarity_p2']
vep2_s = verbs['similarity_eng_p2']
vp2= verbs['similarity_p2'].mean()
vep2= verbs['similarity_eng_p2'].mean()

vp3_s = verbs['similarity_p3']
vep3_s = verbs['similarity_eng_p3']
vp3= verbs['similarity_p3'].mean()
vep3= verbs['similarity_eng_p3'].mean()

vp5_s = verbs['similarity_p5']
vep5_s = verbs['similarity_eng_p5']
vp5= verbs['similarity_p5'].mean()
vep5= verbs['similarity_eng_p5'].mean()



verbs.describe()
```

```python
from scipy.stats import ttest_ind, ttest_rel


# Interpretation    
def ind_t_test(groups, percentage1, percentage2):
    t_stat, p_value = ttest_ind(percentage1, percentage2)
    print(f"T-statistic: {t_stat}")
    print(f"P-value: {p_value}")
    if p_value < 0.05:
        print(f"The difference between the two sets of path similarity percentages between {groups} is statistically significant.\n")
    else:
        print(f"The difference between the two sets of path similarity percentages between {groups} is not statistically significant.\n")

def paired_t_test(groups, percentage1, percentage2):
    t_stat, p_value = ttest_rel(percentage1, percentage2)
    print(f"T-statistic: {t_stat}")
    print(f"P-value: {p_value}")
    if p_value < 0.05:
        print(f"The difference between the two sets of path similarity percentages between {groups} is statistically significant.\n")
    else:
        print(f"The difference between the two sets of path similarity percentages between {groups} is not statistically significant.\n")

groups1 = 'English and Darija in the first prompt'
ind_t_test(groups1,ap1_s,aep1_s)

groups2 = 'English and Darija in the second prompt'
ind_t_test(groups2,ap2_s,aep2_s)

groups3 = 'English and Darija in the third prompt'
ind_t_test(groups3,ap3_s,aep3_s)


groups5 = 'English and Darija in the fifth prompt'
ind_t_test(groups5,ap5_s,aep5_s)



```
